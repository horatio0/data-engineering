## 1. 데이터셋의 개요
제가 선정한 데이터셋은 CNC 밀링 머신 가공 데이터셋입니다. CNC 장비란 컴퓨터 수치 제어 기술을 활용하여 금속, 목재, 플라스틱 등의 재료를 정밀하게 절삭하는 기계입니다. 데이터셋은 CNC 장비가 재료를 가공하는 동안 발생하는 센서 데이터를 수집한 시계열 데이터입니다. 해당 데이터셋은 품질 불량 원인 분석을 위해 사용됩니다.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['axes.unicode_minus'] = False

data_dir = './data/dataset/CNC 비식별화 원본데이터_1209/CNC Virtual Data set _v2'
csv_files = [f for f in os.listdir(data_dir) if f.endswith('.csv')]

sample_df = pd.read_csv(os.path.join(data_dir, 'experiment_01.csv'))
display(sample_df.head())
```

## 2. 데이터셋의 변수 설명
CNC 머신 부품들의 위치, 속도, 가속도, 전압, 전류 등의 데이터가 포함되어 있습니다.

1. 위치/속도/가속도:
   - `_Set[Feature]`: CNC 컨트롤러가 모터에 지시한 목표 값
   - `_Actual[Feature]`: 모터가 이동한 측정 값
2. 전기적 특성:
   - `_CurrentFeedback`, `_DCBusVoltage`, `_OutputPower` 등. 소재의 가공 난이도와 전류 소모량에 관한 데이터
3. **공정 라벨 (Label):**
   - `Machining_Process`: 현재 어떠한 작업이 진행되고 있는지 나타내는 변수

## 3. 데이터셋에 대한 전처리 이슈

1. 벡터화 연산을 통한 파생 변수 생성
   - 센서 데이터 분석의 핵심은 목표과 실제 간의 차이를 구하는 것입니다. 
2. 데이터 결합의 원리
   - 25개의 실험 데이터 파일과 실험별 메타데이터가 분리되어 있습니다. 이 둘을 연결하기 위해서는 공통 키를 기준으로 Join 연산을 수행해야 분석 가능한 하나의 통합된 형태로 가공할 수 있습니다.
3. 스케일링
   - 위치 값과 가속도 오차, 전력 값의 단위가 크게 차이납니다. 머신러닝 관점에서 변수들의 단위 차이가 왜곡을 유발하지 않도록 스케일링이 필요합니다.

```python
sample_df['X_Pos_Error'] = sample_df['X_SetPosition'] - sample_df['X_ActualPosition']
sample_df['Y_Pos_Error'] = sample_df['Y_SetPosition'] - sample_df['Y_ActualPosition']
sample_df['Z_Pos_Error'] = sample_df['Z_SetPosition'] - sample_df['Z_ActualPosition']

display(sample_df['Machining_Process'].value_counts())
```

## 4. EDA: 분포 시각화
위치 오차와 스핀들 파워의 분포 시각화입니다.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(sample_df['X_Pos_Error'], bins=50, kde=True, ax=axes[0], color='blue')
axes[0].set_title('Distribution of X-axis Position Error')
axes[0].set_xlabel('Error Value')

sns.histplot(sample_df['S_OutputPower'], bins=50, kde=True, ax=axes[1], color='orange')
axes[1].set_title('Distribution of Spindle Output Power')
axes[1].set_xlabel('Power')

plt.tight_layout()
plt.show()
```

오차 값이 완벽한 정규분포를 이루기보다는 이상치들을 가질 수 있음을 볼 수 있습니다. 이는 기계가 방향을 전환하거나, 금속에 충돌할 때 오차가 일시적으로 급증하는 CNC 장비의 물리적 관성 한계 때문입니다.

## 5. EDA: 상관관계 (Correlation)
마모 등의 이유로 공구에 저항이 생기면 전류 소모가 늘어나고 위치 오차가 증가할 것이다 라는 가설을 기반으로 계산한 모터의 전력/전류와 위치 오차 간의 피어슨 상관계수입니다.

```python
target_columns = ['X_Pos_Error', 'Y_Pos_Error', 'Z_Pos_Error', 
                  'X_CurrentFeedback', 'Y_CurrentFeedback', 'Z_CurrentFeedback', 'S_OutputPower']

corr_matrix = sample_df[target_columns].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Position Errors and Currents/Power')
plt.show()
```

Z축(드릴이 위아래로 누르는 방향)의 CurrentFeedback과 Pos_Error들이 서로 연결성을 보입니다. 전류 변동은 공구가 금속 표면에 닿을 때 마찰력이 벡터 형태의 힘으로 분산됨을 시사합니다.

## 6. EDA: 그 외의 데이터 탐색 및 시계열 분석
CNC 장비의 특성상 데이터 인덱스가 시간의 흐름을 나타냅니다. 인덱스가 정렬된 상태에서 시간순으로 전류나 오차값이 어떻게 변화하는지 라인플롯으로 그리면 공정의 패턴을 식별할 수 있습니다.

```python
plt.figure(figsize=(14, 4))
plt.plot(sample_df.index, sample_df['S_CurrentFeedback'], label='Spindle Current', color='green', alpha=0.7)
plt.title('Spindle Current Feedback Over Time (Time-series Pattern)')
plt.xlabel('Time Sequence (100ms units)')
plt.ylabel('S_CurrentFeedback')
plt.legend()
plt.show()
```

**분석 및 에세이 결론**

본 과제 과정에서 CNC 밀링 장비의 Raw Data를 탐색하고 가공하며 다각적인 인사이트를 얻을 수 있었습니다. 인덱스 정렬의 원리를 토대로 시간 순으로 변수들을 시각화함으로써 공구의 진입 및 절삭 지점을 분리해낼 수 있었습니다. 그리고 목표값과 실제값의 차이를 벡터화 연산을 통해 파생 변수로 도출하면서 장비의 물리적 한계가 어떻게 데이터 수치로 측정되는지 관찰했습니다. 이러한 분석 파이프라인은 추후 공구의 마모 여부를 판별하는 머신러닝 시스템 설계 시 단순히 모든 피처를 모델에 집어넣는 것이 아니라 오차 피처를 강조하고 시계열 구간에 맞추어 그룹화하는 것이 타당하다는 설계의 근거가 됩니다.
